```python
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.model_selection import ShuffleSplit
from sklearn.metrics import accuracy_score
import numpy as np

# 加載數據
df = pd.read_csv("data.csv")
df.pop('filename')
df.pop('label')
df["tempo"] = df["tempo"].astype(int)

# 初始化 KFold
kf = KFold(n_splits=5, shuffle=True, random_state=42)
SS = ShuffleSplit(n_splits=5, test_size=.25, random_state=0)

kf_result = next(kf.split(df), None)
kf_train = df.iloc[kf_result[0]]
kf_test =  df.iloc[kf_result[1]]
print("K-fold training data\n", kf_train)
print("\nK-fold test data\n", kf_test)


SS_result = next(SS.split(df), None)
SS_train = df.iloc[SS_result[0]]
SS_test =  df.iloc[SS_result[1]]
print("ShuffleSplit training data\n", SS_train)
print("\nShuffleSplit test data\n", SS_test)
```
